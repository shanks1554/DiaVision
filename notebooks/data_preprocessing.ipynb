# Import Data and Libraries

```python
import pandas as pd
```

```python
data=pd.read_csv('../data/diabetes_data.csv')
```

```python
data.head()
```

```python
data.info()
```

# Checking for Missing values

```python
data.isnull().sum()
```

# Droping name column

```python
data=data.drop('Name',axis=1)
```

# Encoding the Data using Label Encoder

Encoding Prakriti

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_prakriti=LabelEncoder()
```

```python
data['Prakriti']=le_prakriti.fit_transform(data['Prakriti'])
```

```python
data['Prakriti'].unique()
```

Encoding Diet

```python
le_diet=LabelEncoder()
```

```python
data['Diet']=le_diet.fit_transform(data['Diet'])
```

```python
data['Diet'].unique()
```

Encoding Food Preferneces

```python
def encode_food_preferance(x):
    if isinstance(x, str):
        prefs=sorted(x.split(', '))
        return ','.join(prefs)
    return ''
```

```python
data['Food Preferences']=data['Food Preferences'].apply(encode_food_preferance)
```

```python
le_food=LabelEncoder()
```

```python
data['Food Preferences']=le_food.fit_transform(data['Food Preferences'])
```

```python
data['Food Preferences']
```

```python
data
```

Encoding Gender

```python
le_gender=LabelEncoder()
```

```python
data['Gender']=le_gender.fit_transform(data['Gender'])
```

```python
data['Gender'].unique()
```

Encoding Sleep Quality

```python
le_sleep_quality=LabelEncoder()
```

```python
data['Sleep Quality']=le_sleep_quality.fit_transform(data['Sleep Quality'])
```

```python
data['Sleep Quality'].unique()
```

Encoding Meal Timings

```python
le_meal_timing=LabelEncoder()
```

```python
data['Meal Timing']=le_meal_timing.fit_transform(data['Meal Timing'])
```

```python
data['Meal Timing'].unique()
```

# Encoding using Binary Encoder

Encoding Excercise

```python
data['Exercise']=data['Exercise'].map({'yes':1,'no':0})
```

```python
data
```

Encoding Daily Routine

```python
data['Daily Routine']=data['Daily Routine'].map({'regular':1, 'irregular':0})
```

```python
data
```

Encoding Diabetes Diagnosis

```python
data['Diabetes Diagnosis']=data['Diabetes Diagnosis'].map({'yes':1,'no':0,'i prefer not to say':0})
```

```python
data
```

Encoding Family History of Diabetes

```python
data['Family History of Diabetes'].unique()
```

```python
data['Family History of Diabetes']=data['Family History of Diabetes'].map({'yes':1,'no':0,"i don't know":0})
```

```python
data
```

# Encoding using Min-Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler
```

Encoding Age

```python
mm_age=MinMaxScaler()
```

```python
data['Age']=mm_age.fit_transform(data[['Age']])
```

```python
data
```

Encoding Stress Level

```python
mm_stress_level=MinMaxScaler()
```

```python
data['Stress Level']=mm_stress_level.fit_transform(data[['Stress Level']])
```

```python
data
```

Encoding Sleep Hours/Night

```python
mm_sleep_hours=MinMaxScaler()
```

```python
data['Sleep Hours/Night']=mm_sleep_hours.fit_transform(data[['Sleep Hours/Night']])
```

```python
data
```
