# Import Libraries & Datasets

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('../data/diabetes_data.csv')
```

```python
data.head()
```

# Gender Distribution (Bar Chart)

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", data=data, palette="pastel")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

# Age Distribution (Histogram)

```python
plt.figure(figsize=(8,5))
sns.histplot(data["Age"], bins=20, kde=True, color="teal")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

# Prakriti Distribution (Pie Chart)

```python
plt.figure(figsize=(6,6))
data["Prakriti"].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Prakriti Distribution")
plt.ylabel("")
plt.show()
```

# Sleep Hours vs. Diabetes (Box Plot)

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Diabetes Diagnosis", y="Sleep Hours/Night", data=data, palette="coolwarm")
plt.title("Sleep Hours vs. Diabetes Diagnosis")
plt.xlabel("Diabetes Diagnosis")
plt.ylabel("Sleep Hours per Night")
plt.show()
```

# Age vs. Diabetes (Box Plot)

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Diabetes Diagnosis", y="Age", data=data, palette="Set2")
plt.title("Age vs. Diabetes Diagnosis")
plt.xlabel("Diabetes Diagnosis")
plt.ylabel("Age")
plt.show()
```

# Exercise Level vs. Diabetes (Stacked Bar Chart)

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Exercise", hue="Diabetes Diagnosis", data=data, palette="Set1")
plt.title("Exercise Level vs. Diabetes Diagnosis")
plt.xlabel("Exercise Level")
plt.ylabel("Count")
plt.legend(title="Diabetes Diagnosis", labels=["No Diabetes", "Diabetes","I prefer not to say"])
plt.show()
```

# Heatmap: Correlation Between Features

```python
encoded_data=pd.read_csv('../data/encoded_data.csv')
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(encoded_data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# Age vs. Sleep Hours (Scatter Plot)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="Sleep Hours/Night", hue="Diabetes Diagnosis", data=data, palette="coolwarm", alpha=0.7)
plt.title("Age vs. Sleep Hours")
plt.xlabel("Age")
plt.ylabel("Sleep Hours per Night")
plt.legend(title="Diabetes Diagnosis")
plt.show()
```

# Stress Level vs. Sleep Quality (Scatter Plot)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Stress Level", y="Sleep Quality", hue="Diabetes Diagnosis", data=data, palette="viridis", alpha=0.7)
plt.title("Stress Level vs. Sleep Quality")
plt.xlabel("Stress Level")
plt.ylabel("Sleep Quality")
plt.legend(title="Diabetes Diagnosis", loc="upper right")
plt.show()
```

# Family History vs. Diabetes (Grouped Bar Chart)

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Family History of Diabetes", hue="Diabetes Diagnosis", data=data, palette="muted")
plt.title("Family History vs. Diabetes Diagnosis")
plt.xlabel("Family History of Diabetes")
plt.ylabel("Count")
plt.legend(title="Diabetes Diagnosis")
plt.show()
```

# Family History Distribution (Pie Chart)

```python
plt.figure(figsize=(6,6))
data["Family History of Diabetes"].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Family History of Diabetes Distribution")
plt.ylabel("")
plt.show()
```

# Diet Type vs. Diabetes (Stacked Bar Chart)

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Diet", hue="Diabetes Diagnosis", data=data, palette="coolwarm")
plt.title("Diet Type vs. Diabetes Diagnosis")
plt.xlabel("Diet Type")
plt.ylabel("Count")
plt.legend(title="Diabetes Diagnosis")
plt.show()
```
